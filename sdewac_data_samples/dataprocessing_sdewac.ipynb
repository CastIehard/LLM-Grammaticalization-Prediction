## FLAN T5: Base

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
def generate_flan_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example prompt
prompt = "Summarize: Albert Einstein was a theoretical physicist who developed the theory of relativity."
print("💬 Prompt:", prompt)
print("✅ Answer:", generate_flan_response(prompt))
```

```python
tasks = [
    "Translate English to German: How are you today?",
    "Classify sentiment: I really hated the food.",
    "Explain: Why do birds fly south in the winter?",
    "Solve math word problem: If Sarah has 3 apples and eats 1, how many are left?"
]

for task in tasks:
    print(f"\n📝 Task: {task}")
    print("💡 Output:", generate_flan_response(task))
```

```python
import os
import pandas as pd

# Set file paths
input_path = "/content/sdewac_10000_lines.tagged"  # replace with your actual path
output_path = "/content/sdewac_cleaned.csv"        # output location

# Prepare containers
data = []
year = None
source = None

# Read and process
with open(input_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()

        if not line:
            continue

        # ✅ Corrected year/source parsing based on actual file format
        if "<year>" in line:
            year = line.replace('<year>="', '').replace('"/>', '').strip()
        elif "<source=" in line:
            source = line.replace('<source="', '').replace('"/>', '').strip()
        elif line.startswith("<") or line.startswith("</"):
            continue  # skip other XML-like markers

        else:
            parts = line.split("\t")
            if len(parts) == 3:
                word, pos, lemma = parts

                # Skip unwanted punctuation/symbol lines
                if not pos.startswith("$"):
                    data.append([word, pos, lemma, year, source])

# Create DataFrame
df = pd.DataFrame(data, columns=["word", "pos", "lemma", "year", "source"])

# Save to CSV
df.to_csv(output_path, index=False, encoding="utf-8")

print(f"✅ Processed data saved to: {output_path}")
```

```python
import os
import pandas as pd
import json

# File paths
tagged_path = "/content/sdewac_10000_lines.tagged"
testset_path = "/content/testset.csv"
sample_csv_path = "/content/test_set_sample.csv"
output_json_path = "/content/sdewac_sampled_output.json"

# Step 1: Load testset
testset_df = pd.read_csv(testset_path)

# Normalize keywords
testset_df["Keyword"] = testset_df["Keyword"].str.strip().str.lower()

# Step 2: Create test sample (10 per score level)
sampled_df = testset_df.groupby("Score", group_keys=False).apply(lambda g: g.sample(n=min(10, len(g)), random_state=42))
sampled_df.to_csv(sample_csv_path, index=False)

# Keyword → score lookup
keyword_score_map = {row["Keyword"]: row["Score"] for _, row in sampled_df.iterrows()}
target_pos_tags = {"APPR", "APPRART", "APPO"}

# Step 3: Process tagged file
results = []
current_sentence = []
current_pos = []

with open(tagged_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()

        if line == "<s>":
            current_sentence = []
            current_pos = []
        elif line == "</s>":
            # Search for keywords in sentence
            for idx, (word, pos_tag) in enumerate(zip(current_sentence, current_pos)):
                word_lc = word.lower()
                if word_lc in keyword_score_map and pos_tag in target_pos_tags:
                    results.append({
                        "sentence": current_sentence,
                        "pos": current_pos,
                        "keyword": word,
                        "grammaticalisation_level": int(keyword_score_map[word_lc])
                    })
                    break
        elif line.startswith("<") or not line:
            continue
        else:
            parts = line.split("\t")
            if len(parts) == 3:
                word, pos_tag, _ = parts
                if not pos_tag.startswith("$"):  # Skip symbols
                    current_sentence.append(word)
                    current_pos.append(pos_tag)

# Step 4: Save to JSON
with open(output_json_path, "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(f"✅ JSON saved to: {output_json_path}")
print(f"✅ Sample test set saved to: {sample_csv_path}")
```

```python
import os
import pandas as pd
import json

# File paths
tagged_path = "/content/sdewac_10000_lines.tagged"
testset_path = "/content/testset.csv"
sample_csv_path = "/content/test_set_sample.csv"
output_json_path = "/content/sdewac_sampled_output_v2.json"

# Step 1: Load testset
testset_df = pd.read_csv(testset_path)

# Normalize keywords
testset_df["Keyword"] = testset_df["Keyword"].str.strip().str.lower()

# Step 2: Create test sample (10 per score level)
sampled_df = testset_df.groupby("Score", group_keys=False).apply(
    lambda g: g.sample(n=min(10, len(g)), random_state=42)
)
sampled_df.to_csv(sample_csv_path, index=False)

# Keyword → score lookup
keyword_score_map = {row["Keyword"]: row["Score"] for _, row in sampled_df.iterrows()}
target_pos_tags = {"APPR", "APPRART", "APPO"}

# Step 3: Process tagged file
results = []
current_sentence = []
current_pos = []

with open(tagged_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()

        if line == "<s>":
            current_sentence = []
            current_pos = []

        elif line == "</s>":
            matched_keywords = []
            levels = []

            for idx, (word, pos_tag) in enumerate(zip(current_sentence, current_pos)):
                word_lc = word.lower()
                if word_lc in keyword_score_map and pos_tag in target_pos_tags:
                    matched_keywords.append((word, idx))
                    levels.append(int(keyword_score_map[word_lc]))

            if matched_keywords:
                results.append({
                    "sentence": current_sentence,
                    "pos": current_pos,
                    "keyword": matched_keywords,
                    "grammaticalisation_level": levels
                })

        elif line.startswith("<") or not line:
            continue
        else:
            parts = line.split("\t")
            if len(parts) == 3:
                word, pos_tag, _ = parts
                if not pos_tag.startswith("$"):  # Skip symbols
                    current_sentence.append(word)
                    current_pos.append(pos_tag)

# Step 4: Save to JSON
with open(output_json_path, "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(f"✅ Enhanced JSON saved to: {output_json_path}")
```
